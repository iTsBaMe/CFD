```python
#solved using explicit method
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
l = 1
deltax = 0.2
deltat = 0.02
x = np.arange(0,1+deltax,deltax)
t = np.arange(0,0.06+deltat,deltat)
n = len(x)
m = len(t)
gamma = 0.5
T = np.zeros((n,m))
boundaryConditions = [0,0]
initialConditions = np.sin(np.pi*x)
T[0,:] = boundaryConditions[0]
T[-1,:] = boundaryConditions[1]
T[:,0] = initialConditions
T.round(4)
```

```python
for j in range(1,m):
    for i in range(1,n-1):
        T[i,j] = gamma*T[i-1,j-1]+(1-2*gamma)*T[i,j-1]+gamma*T[i+1,j-1]
T.round(4)
```

```python
plt.plot(T)
plt.legend(t)
plt.xlabel("Distance")
plt.ylabel("Temperature")
```
